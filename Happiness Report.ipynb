# Bivarate analysis between the target and the features

```python
cols=['Standard Error','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom',
      'Trust (Government Corruption)','Generosity','Dystopia Residual']
```

```python
for a in cols:
    plt.figure(figsize=(10,5))
    sns.regplot(x=a,y='Happiness Score',data=df,color='r')
    plt.show()
```

bservation:

- The graphs shows that Happiness Score is linearly related to Economy (GDP per Capita),Family and Health (Life Expectancy).
- Mild linear relation with Trust (Government Corruption) and Dystopia Residual.
- Almost no linear relation with Standard Error, Generosity

```python
import seaborn as sns
corr = df.corr()
plt.figure(figsize=(10,7))
sns.heatmap(corr,annot=True)
plt.show()
```

Observation:

- We can see that Happiness Score is correlated with Economy (GDP per Capita), Family, Health (Life Expectancy)
- Mild correlated Freedom,Trust (Government Corruption), Dystopia Residualand 
- Not correlated with Standard Error, Generosity

So we not include Standard Error and Generosity in our model buildings

```python
X = df[['Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Dystopia Residual']]
y = df[['Happiness Score']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train)
y_pred = LR.predict(X_test)
```

```python
from sklearn.metrics import r2_score
R2 = r2_score(y_test,y_pred)
R2
```

```python
from sklearn import metrics
MAE= metrics.mean_absolute_error(y_test,y_pred)
MSE= metrics.mean_squared_error(y_test,y_pred)
print("mean absolute error:",MAE)
print("mean squared error:",MSE)
```

```python
from math import sqrt
rmse = sqrt(MSE)
print("Root Mean Squared Error:",rmse)
```
